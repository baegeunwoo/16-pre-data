```python
# 딕셔너리는 어떻게 만들까?
# {key1: value1, key2: value2, ...}

dic = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118' 
}

print(dic)
print(type(dic))
```

```python
# 딕셔너리 쌍 추가하기
a = {1: "a"}
a[2] = 'b'

print(a)
```

```python
a[3] = [1, 2, 3]
print(a)
```

```python
# 딕셔너리 요소 삭제하기
print(a)
del a[1]
print(a)
```

```python
# 딕셔너리에 key를 사용해서 value 얻기
grade = {
	'pey' : 10,
	'juli' : 99
}

print(grade['pey'])
print(grade['juli'])
```

```python
# 딕셔너리 관련 함수
## 1. key 리스트 만들기 - keys

dic = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118' 
}
list(dic.keys())

## 2. key 리스트 만들기 - values
list(dic.values())
```

```python
## 3. key, value 쌍 얻기 - items
print(dic)
dic.items()
list(dic.items())
```

```python
#key, value 쌍 모두 지우기 - clear
a.clear()
print(a)
```

```python
# key로 value 얻기 - get
a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118' 
}
print(a.get('name'))
```

```python
a.get('hobby',0)
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 - in
print('name' in a)
```

```python
#key로 value 꺼내기 - pop
a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118' 
}
result = a.pop("phone")
print(result)
print(a)
```
