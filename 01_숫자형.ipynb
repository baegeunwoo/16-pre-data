```python
# 숫자형은 어떻게 만들고 사용할까?

a = 1234
b = -1234
c = 0

print(a)
print(b)
print(c)

print(type(a))
type(b)
```

```python
# 실수형

a = 1.2
b = -3.14

print(a)
print(b)

print(type(a))
print(type(b))
```

```python
# 8진수와 16진수

a = 0o177
print(a)
```

```python
a = 0x8ff
print(a)
```

```python
# 사칙 연산
a = 3
b = 4

print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
# x의 y 제곱을 나타내는 **연산자

a = 3
b = 4

print(a**b)
```

```python
# 나눗셈 후 나머지는 반환하는 연산자 %

print(7/3)
print(7%3)
```

```python
# 나눗셈 후 몫을 반환하는 연산자 //

print(7//3)
```
